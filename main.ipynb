```python
import pandas as pd
import requests
import lxml
import bs4
import openpyxl
```

```python
res = requests.get("https://www.woundsource.com/trade-name")
soup = bs4.BeautifulSoup(res.text,"lxml")
base_url = "https://www.woundsource.com"
# product links 
product_links = []
for link in soup.find_all("a"):
    url = link.get("href")
    if url and '/product' in url:
        product_links.append(base_url + url)
```

```python
initial = 0

subcategory = []
category = []
product_name = []
info = []
manufacturer = []
about_the_manufacturer = []
manufacturer_contact = []
image_links = []
overview = []
product_features = []
details = []

for i in range(len(product_links)):
    r = requests.get(product_links[initial])
    soup = bs4.BeautifulSoup(r.content, "html.parser")
    
    
    income_product_name_data = soup.find_all("h1",{"class":"page-header"})[0].text
    income_category_data = income_subcategory_data = soup.select(".breadcrumb")[0].text.strip().split('\n')[-2]
    income_subcategory_data = income_subcategory_data = soup.select(".breadcrumb")[0].text.strip().split('\n')[-1]
    income_info_data = soup.find_all("div",{"class":"field-item even","property":"content:encoded"})[0].text.replace("\n","")
    income_manufacturer_data = soup.find_all("div",{"class":"field-item even","property":"dc:title"})[0].text
    income_about_the_manufacturer_data = soup.find_all("div",{"class":"field field-name-body field-type-text-with-summary field-label-hidden margin-top"})[0].text.replace("\n","")
    income_manufacturer_contact_data = soup.find_all("div",{"id":"node-company-company-noderef-group-right","class":"group-right col-md-6"})[0].text.replace('\xa0', ' ')
    
    
    #obtaine for images links
    links = soup.find_all('img')
    count = 0
    for link in links:
        if "https://www.woundsource.com/sites/default/files/styles/medium/public/products/images" in link["src"]:
            image_links.append(link["src"])
            count +=1
    if count == 0:
        image_links.append("no-image")
        count = 0
    
    #obtaine for all product features insert a find list
    x = soup.select("table",{"class":"features-table product-features-table table-bordered table-hover table table-hover table-striped"})
    features = str(x).split("<tr><td>")
    find = []
    for j in features:
        if "ticked" in j:
            a = 1
            for i in j:
                if i == "<":
                    break
                item = j[0:a]
                a += 1
            find.append(item)
    
    
    try:
        overview.append(soup.find_all("div",{"id":"overview","class":"tab-pane active"})[0].text.replace("\n",""))
    except:
        overview.append("no-overview")
    
    
    
    
    try:
        x = soup.find_all("div",{"id":"details","class":"tab-pane"})
        details.append(x[0].text.replace("\n",""))
    except:
        details.append("no-detail")
    
    
    
    
    
    product_name.append(income_product_name_data)
    category.append(income_category_data)
    subcategory.append(income_subcategory_data)
    info.append(income_info_data)
    manufacturer.append(income_manufacturer_data)
    about_the_manufacturer.append(income_about_the_manufacturer_data)
    manufacturer_contact.append(income_manufacturer_contact_data)
    product_features.append(find)
    

    
    initial+=1
    
    
    
    
    
```

```python
data = {"PRODUCT_NAME":product_name,
        "CATEGORY":category,
        "SUBCATEGORY":subcategory,
        "INFO":info,
        "IMAGE_LINKS":image_links,
        "MANUFACTURER":manufacturer,
        "ABOUT_THE_MANUFACTURER":about_the_manufacturer,
        "MANUFACTURER_CONTACT":manufacturer_contact,
        "OVERVIEW":overview,
        "PRODUCT_FEATURES":product_features,
        "DETAILS":details}

df = pd.DataFrame(data)
df
```
